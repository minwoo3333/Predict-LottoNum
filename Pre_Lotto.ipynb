```python
documents = collection.find({}, {'_id': 0}).sort('date', 1)
# 데이터에서 로또 숫자와 보너스 숫자만 추출
data = []

for document in documents:
    lotto_numbers = [
        document.get('ball1'),
        document.get('ball2'),
        document.get('ball3'),
        document.get('ball4'),
        document.get('ball5'),
        document.get('ball6')
        # document.get('bonusball')
    ]
    data.append(lotto_numbers)

data = np.array(data)
```

```python
len(data)
```

```python
# 시계열 데이터 생성
def create_sequences(data, sequence_length):
    X = []
    y = []
    for i in range(len(data) - sequence_length):
        X.append(data[i:i + sequence_length])
        y.append(data[i + sequence_length])
    return np.array(X), np.array(y)
# ex) 1회차 -> 2회차, 2회차 -> 3회차, 3회차->4회차

# 시퀀스 길이 설정
sequence_length = 1

X, y = create_sequences(data, sequence_length)
train_size = int(len(X) * 0.9)
X_train, X_test = X[:train_size], X[train_size:]
y_train, y_test = y[:train_size], y[train_size:]
# 학습용 데이타와 검증용 데이타(최신) 나누기.

print("입력 데이터(X):")
print(X_train)
print("결과값(y):")
print(y_train)
```

## RNN(순환 신경망)
- 과거의 정보를 사용하여 현재 및 미래의 입력에 대한 신경망의 성능을 개선
- 루프 구조를 통해 신경망은 은닉 상태에 과거의 정보를 저장하고 시퀀스에 대해 연산이 가능

- 역전파(backprogation) 과정에서 기울기 소실 or 폭발이 발생.

# LSTM(Long Short-Term Memory)
- LSTM(Long Short-Term Memory)은 순환 신경망(RNN, Recurrent Neural Network)의 한 종류로 시퀀스 데이터를 처리하는 데 주로 사용
- 각 LSTM 셀은 입력 데이터의 시퀀스를 받아들이고, 이전 시간 단계의 상태를 기억하는 능력을 갖춤
- 기억할 것과 잊을 것을 선택해 중요한 정보만 기억하는 것이 핵심(hidden layer의 메모리 셀에 입력 게이트, 망각 게이트, 출력 게이트를 추가하여 불필요한 기억을 지우고, 기억해야할 것들을 정함)

✔ 시퀀스 데이터: 일련의 데이터 포인트가 순서대로 배열된 데이터(ex: 시계열, 텍스트, 음성, DNA 데이터)

### LSTM 파라미터
- input_dim: 입력 데이터의 특성 수
- output_dim: LSTM 층의 출력 차원, 즉 LSTM 셀의 유닛 수(윗 그림에서 A)
- activation: LSTM 셀 내부의 활성화 함수
- return_sequences=True : 각 타임스텝의 출력을 다음 LSTM 레이어에 전달할 수 있어 모델이 더 깊은 시퀀스 패턴을 학습할 수 있다

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout
# Dense : 완전히 연결된(fully connected) 뉴런 레이어를 정의할 때 사용
from tensorflow.keras.optimizers import Adam
from sklearn.model_selection import train_test_split
```

```python
# LSTM 모델 정의
model = Sequential()
model.add(LSTM(100, return_sequences=True, activation='tanh', input_shape=(sequence_length, 6)))
model.add(Dropout(0.2))    # 과적합을 방지하기 위해 학습 중에 랜덤하게 뉴런의 20%를 비활성화
model.add(LSTM(100, activation='tanh')) # 두 번째 LSTM 레이어 추가로 이전 LSTM 레이어의 출력(시퀀스)을 입력으로 받음
model.add(Dense(6)) #출력시 차원을 6으로 지정

optimizer = Adam(learning_rate=0.001)
model.compile(optimizer=optimizer, loss='mse')

# 모델 훈련
model.fit(X_train, y_train, epochs=200, verbose=1, validation_data=(X_test, y_test))
```

```python
# 모델 평가
loss = model.evaluate(X_test, y_test, verbose=0)
print(f'Test Loss: {loss}')
```

```python
# 모델 예측
y_pred = model.predict([[[6, 14, 25, 33, 40, 44]]])
y_pred = np.round(y_pred).astype(int)
y_pred = np.clip(y_pred, 1, 45)

print("예측 값:")
print(y_pred)
```
